```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as pit
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/facespn.bmp')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
kernel = np.ones((3,3), np.uint8)
erosion = cv2.erode(gimg1, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
opening = cv2.dilate(gimg1, kernel, iterations=1)
cv2_imshow(opening)
```

```python
dilate2 = cv2.dilate(opening, kernel, iterations=2)
cv2_imshow(dilate2)
```

```python
erode2 = cv2.erode(dilate2, kernel, iterations=2)
cv2_imshow(erode2)
```
