```python
import numpy as np
gPlus = np.zeros((256, 256), dtype=np.ubyte)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    pValue= np.int32(gimg1[h,w]) + np.int32(gimg2[h,w])
    if pValue > 255:
      gPlus[h,w] = 255;
    else:
      gPlus[h,w] = pValue;

plt.imshow (gPlus, cmap= 'gray')
```
