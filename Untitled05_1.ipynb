```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
def saturation(value):
  if value > 255:
    return 255
  elif (value < 0):
    return 0
  else:
    return value
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel5 = np.ones((5,5), np.float32)/25
kernel7 = np.ones((7,7), np.float32)/49
```

```python
output = cv2.filter2D(gimg1, -1, kernel3)
cv2_imshow(output)
```

```python
output = cv2.filter2D(gimg1, -1, kernel5)
cv2_imshow(output)
```

```python
output = cv2.filter2D(gimg1, -1, kernel7)
cv2_imshow(output)
```

```python
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = 8/9
kernel3[1][1] = -1/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```

```python
output = cv2.filter2D(gimg1, -1, kernel7)
# cv2_imshow(output)
output = output * 5
pit.imshow (output, cmap= 'gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/Lena-g1.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
def saturation(value):
  if value > 255:
    return 255
  elif (value < 0):
    return 0
  else:
    return value

gOut = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.float32) # Ensure gOut is initialized as float32

for h in range(1, gimg1.shape[0]-1):
  for w in range(1, gimg1.shape[1]-1):
    sum = 0
    for i in range(-1, 2):
      for j in range(-1, 2):
          sum += gimg1[h+i][w+j] * kernel3[i+1][j+1]
    sum = saturation(sum)
    gOut[h][w] = sum

cv2_imshow(gOut.astype(np.ubyte)) # Cast back to ubyte for displaying
```
